```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
lamda = np.array([[1.0, 60.0, 10.0],
                [1.0, 70.0, 8.0],
                [1.0, 75.0, 5.0],
                [1.0, 55.0, 12.0],
                [1.0, 80.0, 6.0],
                [1.0, 65.0, 9.0],
                [1.0, 68.0, 7.0],
                [1.0, 62.0, 11.0],
                [1.0, 72.0, 4.0],
                [1.0, 67.0, 8.0]])

q = np.array([[100.0], [120.0], [110.0], [90.0], [130.0], [105.0], [115.0], [95.0], [125.0], [112.0]])

lamda_trans = lamda.transpose()
tao = np.matmul(lamda_trans, lamda)
tao_inv = np.linalg.inv(tao)
lamda_transTimesQ = np.matmul(lamda_trans, q)
p = np.matmul(tao_inv, lamda_transTimesQ)
#print(p[0],'\n',p[1],'\n',p[2])

fig = plt.figure()
ax = fig.add_subplot(projection = "3d")
ax.scatter(lamda[:, 1], lamda[:, 2], q)

x = np.linspace(min(lamda[:, 1]), max(lamda[:, 1]), 100)
y = np.linspace(min(lamda[:, 2]), max(lamda[:, 2]), 100)
X,Y = np.meshgrid(x, y)

Z = p[0] + p[1] * X + p[2] * Y

ax.plot_surface(X, Y, Z, alpha = 0.5)

ax.set_xlabel("Humidity")
ax.set_ylabel("Wind Speed")
ax.set_zlabel("Energy Consumption")

print("Energy consumption at x = 50 and y = 100:")
Z = p[0] + p[1] * 50 + p[2] * 100
print(Z[0])
```
