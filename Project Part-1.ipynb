```python
import numpy as np
import matplotlib.pyplot as plt
P1f=389/1000
P2f=151/611
P3f=7/10
P1=7/20
P2=11/50
P3=457/1000
M12f=23/100
M23f=2/5
M12=4/25
M23=37/100
F2m=1
F2f=2
m1=np.array([P1,0,0,0,F2m,0])
m2=np.array([M12,P2,0,0,0,0])
m3=np.array([0,M23,P3,0,0,0])
f1=np.array([0,0,0,P1f,F2f,0])
f2=np.array([0,0,0,M12f,P2f,0])
f3=np.array([0,0,0,0,M23f,P3f])
L=(m1,
  m2,
  m3,
  f1,
  f2,
  f3)
Zk=np.array([[50],
             [50],
             [50],
             [50],
             [50],
             [50]])
for i in range(50):
    Zn=np.matmul(L,Zk)
print(Zn)
```

```python
male_age_group=[[],[],[]]
female_age_group=[[],[],[]]
for i in range(51): 
    
    #51 is used as we need to find population in 50th year
    male_age_group[0].append(np.log(Zk[0][0]))
    male_age_group[1].append(np.log(Zk[1][0]))
    male_age_group[2].append(np.log(Zk[2][0]))
    female_age_group[0].append(np.log(Zk[3][0]))
    female_age_group[1].append(np.log(Zk[4][0]))
    female_age_group[2].append(np.log(Zk[5][0]))
    
    Zk=np.matmul(L,Zk)
print(Zk)
years=list(range(51))

plt.figure(figsize=(10,7))

  

for i in range (3):
    plt.plot(years,male_age_group[i],label=f'Male class{i+1}')
for i in range (3):
     plt.plot(years,female_age_group[i],label=f'Female class{i+1}')
        
        
                      
                      #if i is 0, the label would be 'Male class1'. If i is 1, the label would be 'Male class2', and so on
plt.xlabel('years')
plt.ylabel('population')
plt.title('population and time')
plt.legend()
plt.grid(True)
plt.show()
print(years)
```
