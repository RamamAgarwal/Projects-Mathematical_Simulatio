```python
import random
import matplotlib.pyplot as plt
n=int(input("Enter the number of trials: "))
count=0;
trials=[]
heads=[]
for i in range(n+1):
    s=random.randint(0,1) 
    trials.append(i+1) #Appending number of coin flips
    if(s==0):
        count+=1 #Total number of heads
    heads.append(count) #Appending number of heads
print(f"Number of heads:{count}")
plt.xlabel("Number of coin flips")
plt.ylabel("Frequency of heads")
plt.plot(trials,heads)
plt.show()
```

```python
l=[]
ls=[]
p,q=0,0
for i in range(1,n):
    x=random.randint(0,1)
    if x==0:
        if i==1:
            p+=1 #For the first flip, incrementing for first head
        else:
            q=(p/(i-1)) #This is for calculating the previous frequency
            p+=1
    l.append(abs(p/i-q)) #Appending the (current frequency of heads - previous frequency of heads) and used abs because probability cannot be negative
    ls.append(i) #Appending number of coin flips
plt.plot(ls,l)
plt.xlabel('Number of coin flips')
plt.ylabel('Fluctuations f(H)')
```
