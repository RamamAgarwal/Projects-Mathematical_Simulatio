```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d

lamda = np.array([[1.0, 60.0, 10.0],
                [1.0, 70.0, 8.0],
                [1.0, 75.0, 5.0],
                [1.0, 55.0, 12.0],
                [1.0, 80.0, 6.0],
                [1.0, 65.0, 9.0],
                [1.0, 68.0, 7.0],
                [1.0, 62.0, 11.0],
                [1.0, 72.0, 4.0],
                [1.0, 67.0, 8.0]])

q = np.array([[25.0], [15.0], [20.0], [28.0], [10.0], [22.0], [18.0], [26.0], [17.0], [23.0]])

lamda_trans = lamda.transpose()
ta = np.matmul(lamda_trans, lamda)
ta_inv = np.linalg.inv(ta)
lamda_transTimesQ = np.matmul(lamda_trans, q)
p = np.matmul(ta_inv, lamda_transTimesQ)
print(p[0],'\n',p[1],'\n',p[2])

fig = plt.figure()
ax = fig.add_subplot(projection = "3d")
ax.scatter(lamda[:, 1], lamda[:, 2], q)

x = np.linspace(min(lamda[:, 1]), max(lamda[:, 1]), 100)
y = np.linspace(min(lamda[:, 2]), max(lamda[:, 2]), 100)
X,Y = np.meshgrid(x, y)
Z = p[0] + p[1] * X + p[2] * Y

ax.plot_surface(X, Y, Z, alpha = 0.5)

ax.set_xlabel("Humidity")
ax.set_ylabel("Wind Speed")
ax.set_zlabel("Temperature")

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d

arr = np.array([[1.0, 60.0, 10.0],
                [1.0, 70.0, 8.0],
                [1.0, 75.0, 5.0],
                [1.0, 55.0, 12.0],
                [1.0, 80.0, 6.0],
                [1.0, 65.0, 9.0],
                [1.0, 68.0, 7.0],
                [1.0, 62.0, 11.0],
                [1.0, 72.0, 4.0],
                [1.0, 67.0, 8.0]])

q = np.array([[25.0], [15.0], [20.0], [28.0], [10.0], [22.0], [18.0], [26.0], [17.0], [23.0]])

arr_transpose = arr.transpose()
lam = np.matmul(arr_transpose, arr)
print(lam)
lam_inverse = np.linalg.inv(lam)
#print(lam_inverse)
lamInvTimesQ = np.matmul(arr_transpose, q)
p = np.matmul(lam_inverse, lamInvTimesQ)
print(p[0],'\n',p[1],'\n',p[2])

ax = plt.figure().add_subplot(projection='3d')
ax.scatter(arr[:, 1], arr[:, 2], q, color='r', marker='o', label='Original Data')

x = np.linspace(min(arr[:, 1]), max(arr[:, 1]), 10)
y = np.linspace(min(arr[:, 2]), max(arr[:, 2]), 10)
X, Y = np.meshgrid(x, y)
Z = p[0] + p[1]*X + p[2]*Y

ax.plot_surface(X, Y, Z, alpha=0.3)

ax.set_xlabel('Humidity')
ax.set_ylabel('Wind Speed')
ax.set_zlabel('Temperature')
plt.legend()
plt.show()
```
