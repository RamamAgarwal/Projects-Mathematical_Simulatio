```python
import numpy as np
import math as m
import matplotlib.pyplot as plt

def Gaussian(x, mu, sigma):
    P = []
    for i in x:
        p_i = 1 / (m.sqrt(2 * m.pi) * sigma) * m.exp(-((i - mu) ** 2) / (2 * sigma ** 2))
        P.append(p_i)
    return P
        
mu = 0.2
x = np.linspace(-50, 50, 100)
sigma = np.array([1, 2, 4])

for j in range(len(sigma)):
    PMF = Gaussian(x, mu, sigma[j])
    plt.plot(x, PMF, label=f"sigma = {sigma[j]}")

plt.xlabel('X')
plt.ylabel('P(X)')
plt.legend()
plt.show()
```
