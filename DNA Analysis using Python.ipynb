```python
def rev(s):
    r = s[::-1]
    l = []
    for i in r:
        if i == "A":
            l.append("T")
        elif i == "C":
            l.append("G")
        elif i == "G":
            l.append("C")
        elif i == "T":
            l.append("A")
    return ''.join(l)

file = open("C:\\Users\\ramam\\Downloads\\seq.fa", 'r')
d = ""
for line in file:
    d += line.strip()
d2 = d.split(">seq")
c=0
for j in d2[1:]:
    if j[0].isdigit():
        res = rev(j)
        c+=1
        print(f"Reverse complement of Sequence{c}:")
        print(f'{res}\n')
```

```python
secount=0
def count(a):
    global secount
    n = {"A": 0, "C": 0, "G": 0, "T": 0}
    for i in a:
        if i == "A":
            n['A'] += 1
        if i == "C":
            n['C'] += 1
        if i == "G":
            n['G'] += 1
        if i == "T":
            n['T'] += 1
    secount+=1
    print(f'in seq{secount}')
    for i, k in n.items():
        print(f' {i} {k}')
def tcount(a):
    n = {"A": 0, "C": 0, "G": 0, "T": 0}
    for i in a:
        if i == "A":
            n['A'] += 1
        if i == "C":
            n['C'] += 1
        if i == "G":
            n['G'] += 1
        if i == "T":
            n['T'] += 1
            
    print("total count")
    for i, k in n.items():
        print(f' {i} {k}')

file = open("C:\\Users\\ramam\\Downloads\\seq.fa", 'r')
data = ""
for line in file:
    data += line.strip()
        
tcount(data)
print(".....................................................\n")

data2 = data.split(">seq")

for sequence in data2[1:]:
    count(sequence)
```

```python
def sp(p_sp, p_nsp, p_w_sp, p_w_nsp):
    total_word_prob = p_w_sp * p_sp + p_w_nsp * p_nsp
    return (p_w_sp * p_sp) / total_word_prob

p_spam = float(input("Enter probability for a message to be spam (P(S)): "))
p_not_spam = 1 - p_spam  # Law of Negation
p_word_given_spam = float(input("Enter P(B|S): "))
p_word_given_not_spam = float(input("Enter P(B|NS): "))
spam_prob = sp(p_spam, p_not_spam, p_word_given_spam, p_word_given_not_spam)

print(f"Probability of a message containing the word being spam: {spam_prob}")
```
