```python
import random
import matplotlib.pyplot as plt
n=int(input("Enter the number of trials: "))
count=0;
trials = []
heads=[]
plt.figure(figsize=(8,6))
for i in range(n+1):
    s=random.randint(0,1)
    trials.append(i+1)
    if(s==0):
        count+=1
    heads.append(count)
print(count)
plt.xlabel("Number of coin flips")
plt.ylabel("Frequency of heads")
plt.plot(trials,heads)
plt.show()
```

```python
plt.figure(figsize=(8,6))
ls=[]
count=0
trials1=[]
for i in range(n+1):
    s=random.randint(0,1)
    if(s==0):
        count+=1
    ls.append(count/(i+1))
    trials1.append(i+1)
plt.plot(trials1,ls)
print(ls)
plt.show()
```

```python
fluctuations = [0]  
for i in range(1, len(heads)):
    fluctuations.append((heads[i] / trials[i]) ** 0.5)
plt.plot(trials, fluctuations, label='Fluctuations')

plt.xlabel("Number of coin flips")
plt.ylabel("Frequency of heads / Fluctuations")
plt.legend()
plt.show()
```
