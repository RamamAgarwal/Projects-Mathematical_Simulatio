```python
#3
def rev(s):
    r = s[::-1]
    l = []
    for i in r:
        if i == "A":
            l.append("T")
        elif i == "C":
            l.append("G")
        elif i == "G":
            l.append("C")
        elif i == "T":
            l.append("A")
    return ''.join(l)

file = open("C:\\Users\\ramam\\Downloads\\seq.fa", 'r')
d = ""
for line in file:
    d += line.strip()
d2 = d.split(">seq")
c=0
for j in d2[1:]:
    if j and j[0].isdigit():
        res = rev(j)
        c+=1
        print(f"Reverse complement of Sequence{c}:")
        print(f'{res}\n')
```

```python
#4
def PatternCount(Text, Pattern):
    count = 0
    for i in range(len(Text)-len(Pattern)+1):
        if Text[i:i+len(Pattern)] == Pattern: 
            count = count+1
    return count
def FrequencyMap(Text, k):
    freq = {}
    n = len(Text)
    for i in range(n - k + 1):
        Pattern = Text[i:i + k]
        if Pattern in freq:
            freq[Pattern] += 1
        else:
            freq[Pattern] = 1
    return freq
def FrequentWords(Text, k): 
    words = [] 
    freq = FrequencyMap(Text, k) 
    m = max(freq.values()) 
    for key in freq:
        if freq[key] == m:
            words.append(key)
    return words
Text=input("Enter the DNA sequence: ")
k=int(input("Enter the value of k:"))
w=FrequentWords(Text,k)
n=PatternCount(Text,w[0])
print(f'The most frequent kmer in the given sequence is {w[0]} and its frequency is {n}')
```
